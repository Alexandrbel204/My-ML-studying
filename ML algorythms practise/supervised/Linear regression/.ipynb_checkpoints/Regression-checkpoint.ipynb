# Linear Regression

#### First is matrix method -> w = (X^T*X)^-1*X^T*y

```python
class MatrixLinearRegression:
    def fit(self, X, y):
        X = np.insert(X, 0, 1, axis=1)
        XT_X_INV = np.linalg.inv(X.T @ X) # (X.T * X) ** (-1) inverse matrix
        weights = np.linalg.multi_dot([XT_X_INV, X.T, y])
        self.bias, self.weights = weights[0], weights[1:]
        
    def predict(self, X_test):
        return X_test @ self.weights + self.bias
```

#### Second is gradients descent

```python
class GDLinearRegression:
    def __init__(self, learning_rate=0.01, tolerance=1e-8):
        self.learning_rate = learning_rate
        self.tolerance = tolerance

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.bias, self.weights = 0, np.zeros(n_features)
        previous_db, previous_dw = 0, np.zeros(n_features)
        
        while True:
            y_pred = X @ self.weights + self.bias
            db = 1/n_samples * np.sum(y_pred-y)
            dw = 1/n_samples * X.T @ (y_pred-y)
            self.bias -= self.learning_rate * db
            self.weights -= self.learning_rate * dw

            abs_db = np.abs(db-previous_db)
            abs_dw = np.abs(dw-previous_dw)

            if abs_db < self.tolerance:
                if abs_dw.all() < self.tolerance:
                    break

            previous_db = db
            previous_dw = dw

            

    def predict(self, X_test):
        return X_test @ self.weights + self.bias
```

#### Linear regression (matrix method)

```python
matrix_linear_regression = MatrixLinearRegression()
matrix_linear_regression.fit(X1_train_s, y1_train)
matrix_lr_pred_res = matrix_linear_regression.predict(X1_test_s)
matrix_lr_r2 = r2_score(y1_test, matrix_lr_pred_res)
matrix_lr_mape = mean_absolute_percentage_error(y1_test, matrix_lr_pred_res)
```

```python
print(f'Matrix Linear regression  R2 score: {matrix_lr_r2}')
print(f'Matrix Linear regression MAPE: {matrix_lr_mape}', '\n')

print(f'weights: {matrix_linear_regression.bias, *matrix_linear_regression.weights}')
print(f'prediction: {matrix_lr_pred_res}')
```

#### Linear Regression (Gradient)

```python
gradient_linear_regression = GDLinearRegression()
gradient_linear_regression.fit(X1_train_s, y1_train)
gradient_lr_pred_res = gradient_linear_regression.predict(X1_test_s)
gradient_lr_r2 = r2_score(y1_test, gradient_lr_pred_res)
gradient_lr_mape = mean_absolute_percentage_error(y1_test, gradient_lr_pred_res)
```

```python
print(f'Linear regression R2 score: {gradient_lr_r2}')
print(f'Linear regression MAPE: {gradient_lr_mape}', '\n')

print(f'weights: {gradient_linear_regression.bias, *gradient_linear_regression.weights}')
print(f'prediction: {gradient_lr_pred_res}')
```

#### Linear regression (scikit-learn)

```python
sk_linear_regression = LinearRegression()
sk_linear_regression.fit(X1_train, y1_train)

sk_lr_pred_res = sk_linear_regression.predict(X1_test)
sk_lr_r2 = r2_score(y1_test, sk_lr_pred_res)
sk_lr_mape = mean_absolute_percentage_error(y1_test, sk_lr_pred_res)

print(f'Scikit-learn Linear regression R2 score: {sk_lr_r2}')
print(f'Scikit-learn Linear regression MAPE: {sk_lr_mape}', '\n')

print(f'weights: {sk_linear_regression.intercept_, *sk_linear_regression.coef_}')
print(f'prediction: {sk_lr_pred_res}', '\n')

sk_linear_regression.fit(X1_train_s, y1_train)
print(f'scaled weights: {sk_linear_regression.intercept_, *sk_linear_regression.coef_}')
```

# Polynomial (multiple) linear regression

Линейную регрессию также можно применять к данным с нелинейной зависимостью, добавив степени каждого признака в виде новых признаков с последующим обучением на полученном датасете. Такой подход позволяет улавливать линейные связи в многомерном пространстве признаков и называется полиномиальной регрессией, а для преобразования признаков в полином степени n в scikit-learn имеется класс PolynomialFeatures, который кроме степеней каждого признака ещё добавляет их комбинации до заданной степени. Например, для признаков a и b со степенью 3 кроме a^2, a^3 и b^2, b^3 будут также добавлены их комбинации в виде ab, a^2b  и ab^2.

Стоит отметить, что полиномиальная регрессия всё ещё остаётся линейной, однако на графике в исходном пространстве признаков она будет выглядеть в виде кривой, поскольку полученная гиперплоскость в многомерном пространстве признаков будет соответствовать сложной кривой линии в исходном пространстве.

```python
from itertools import combinations_with_replacement

def polynomial_features(X, degree):
    n_samples, n_features = X.shape
    # Список всех возможных комбинаций признаков с повторениями (например, (0,0), (0,1), (1,1), ...)
    combinations = list(combinations_with_replacement(range(n_features), degree))
    
    def compute_feature(x, comb):
        return np.prod([x[i] for i in comb])

    # Начинаем с bias term
    X_poly = [np.ones(n_samples)]
    
    for d in range(1, degree + 1):
        for comb in combinations_with_replacement(range(n_features), d):
            X_poly.append(np.prod(X[:, comb], axis=1))
    
    return np.vstack(X_poly).T
```

```python
class PolynomialMatrixLinearRegression:
    def __init__(self, degree):
        self.degree = degree

    def fit(self, X, y):
        X_poly = polynomial_features(X, self.degree)
        XT_X_INV = np.linalg.inv(X_poly.T @ X_poly)
        self.weights = XT_X_INV @ X_poly.T @ y

    def predict(self, X_test):
        X_poly = polynomial_features(X_test, self.degree)
        return X_poly @ self.weights
```

```python
class PolynomialGDLinearRegression:
    def __init__(self, degree, learning_rate=0.01, tolerance=1e-8, max_iter=10000):
        self.degree = degree
        self.learning_rate = learning_rate
        self.tolerance = tolerance
        self.max_iter = max_iter

    def fit(self, X, y):
        X_poly = polynomial_features(X, self.degree)
        n_samples, n_features = X_poly.shape
        self.weights = np.zeros(n_features)

        for _ in range(self.max_iter):
            y_pred = X_poly @ self.weights
            gradient = 1/n_samples * X_poly.T @ (y_pred - y)
            new_weights = self.weights - self.learning_rate * gradient

            if np.all(np.abs(new_weights - self.weights) < self.tolerance):
                break

            self.weights = new_weights

    def predict(self, X_test):
        X_poly = polynomial_features(X_test, self.degree)
        return X_poly @ self.weights
```

```python
df = pd.read_csv("Ice_cream selling data.csv")
X = df[['Temperature (°C)']].values
y = df['Ice Cream Sales (units)'].values

# Параметры 
degree = 2

# Sklearn модель
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)
sk_model = LinearRegression()
sk_model.fit(X_poly, y)
sk_preds = sk_model.predict(X_poly)

# Моя модель 
my_model = PolynomialMatrixLinearRegression(degree=degree)
my_model.fit(X, y)
my_preds = my_model.predict(X)

plt.figure(figsize=(10,6))
plt.scatter(X, y, color='black', label='Actual Data')
plt.plot(X, sk_preds, color='blue', linestyle='--', label='Sklearn Model')
plt.plot(X, my_preds, color='green', linestyle='-', label='Custom Model')
plt.title('Polynomial Regression Comparison (Degree {})'.format(degree))
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("SKLEARN MSE:", mean_squared_error(y, sk_preds))
print("CUSTOM MSE :", mean_squared_error(y, my_preds))
print("R² SKLEARN :", r2_score(y, sk_preds))
print("R² CUSTOM  :", r2_score(y, my_preds))
```

```python
#  GD 
gd_model = PolynomialGDLinearRegression(degree=degree, learning_rate=0.0001)
gd_model.fit(X, y)
gd_preds = gd_model.predict(X)

# График 
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='black', label='Actual Data')
plt.plot(X, sk_preds, color='blue', linestyle='--', label='Sklearn Model')
plt.plot(X, my_preds, color='green', linestyle='-', label='Matrix Model')
plt.plot(X, gd_preds, color='red', linestyle='-.', label='GD Model')
plt.title('Polynomial Regression Comparison (Degree {})'.format(degree))
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# 
from sklearn.metrics import mean_squared_error
print("GD MSE:", mean_squared_error(y, gd_preds))
```

# Lasso/Ridge regression

Если в полиномиальной регрессии регуляризация позволяет уменьшить кривизну линии через снижение количества полиномиальных степеней, то в случае с линейной регрессией регуляризация будет заключаться в изменении наклона линии путём ограничения весов модели, обменивая более высокое смещение на низкую дисперсию.

## Гребневая регрессия (Ridge regression) или регуляризация Тихонова 
применяется в случае мультиколлинеарности через добавление L2-регуляризации к функции потерь во время обучения и сильнее всего занижает веса для признаков с высокой корреляцией: их значения будут приближаться к нулю, но никогда его не достигнут. Лучше всего применять гребневую регрессию после стандартизации признаков.

```python
class PolynomialRidgeRegression:
    def __init__(self, degree, alpha=1.0):
        self.degree = degree
        self.alpha = alpha

    def fit(self, X, y):
        X_poly = polynomial_features(X, self.degree)
        n_features = X_poly.shape[1]
        identity = np.eye(n_features)
        identity[0, 0] = 0  # не регуляризуем bias
        self.weights = np.linalg.inv(X_poly.T @ X_poly + self.alpha * identity) @ X_poly.T @ y

    def predict(self, X):
        X_poly = polynomial_features(X, self.degree)
        return X_poly @ self.weights

class PolynomialRidgeGDRegression:
    def __init__(self, degree, alpha=1.0, learning_rate=0.001, tolerance=1e-6, max_iter=10000):
        self.degree = degree
        self.alpha = alpha
        self.learning_rate = learning_rate
        self.tolerance = tolerance
        self.max_iter = max_iter

    def fit(self, X, y):
        X_poly = polynomial_features(X, self.degree)
        n_samples, n_features = X_poly.shape
        self.weights = np.zeros(n_features)

        for _ in range(self.max_iter):
            y_pred = X_poly @ self.weights
            error = y_pred - y

            # L2 regularization
            gradient = (1/n_samples) * X_poly.T @ error + self.alpha * np.r_[0, self.weights[1:]]  # no reg for bias
            new_weights = self.weights - self.learning_rate * gradient

            if np.all(np.abs(new_weights - self.weights) < self.tolerance):
                break

            self.weights = new_weights

    def predict(self, X):
        X_poly = polynomial_features(X, self.degree)
        return X_poly @ self.weights
```

## Lasso regression (L1)

обычно используется для отбора признаков через добавление L1-регуляризации к функции потерь во время обучения. Проще говоря, лассо-регрессия стремится уменьшить число параметров модели путем зануления весов для неинформативных и избыточных признаков, что на выходе даст разреженную модель (с небольшим числом ненулевых весов признаков).

Lasso использует L1 регуляризацию, у которой градиент не дифференцируем в 0. Для Lasso нужен либо:

Subgradient метод, либо

Coordinate Descent (что делает sklearn)

Но мы можем реализовать простой subgradient метод (чуть менее точный, но работает):

```python
class PolynomialLassoGDRegression:
    def __init__(self, degree, alpha=1.0, learning_rate=0.001, tolerance=1e-6, max_iter=10000):
        self.degree = degree
        self.alpha = alpha
        self.learning_rate = learning_rate
        self.tolerance = tolerance
        self.max_iter = max_iter

    def fit(self, X, y):
        X_poly = polynomial_features(X, self.degree)
        n_samples, n_features = X_poly.shape
        self.weights = np.zeros(n_features)

        for _ in range(self.max_iter):
            y_pred = X_poly @ self.weights
            error = y_pred - y

            # L1 gradient (subgradient)
            l1_grad = np.sign(self.weights)
            l1_grad[0] = 0  # don't regularize bias

            gradient = (1/n_samples) * X_poly.T @ error + self.alpha * l1_grad
            new_weights = self.weights - self.learning_rate * gradient

            if np.all(np.abs(new_weights - self.weights) < self.tolerance):
                break

            self.weights = new_weights

    def predict(self, X):
        X_poly = polynomial_features(X, self.degree)
        return X_poly @ self.weights

```

## ElasticNet = Ridge + Lasso

представляет собой комбинацию L1 и L2-регуляризаций через отношение их смеси r, что может принести особую пользу в ситуациях, когда в данных необходимо одновременно выполнять отбор признаков и бороться с мультиколлинеарностью. В Scikit-Learn для управления смесью Ridge и Lasso используется "l1_ratio".

```python
class PolynomialElasticNetGDRegression:
    def __init__(self, degree, alpha=1.0, l1_ratio=0.5, learning_rate=0.001, tolerance=1e-6, max_iter=10000):
        self.degree = degree
        self.alpha = alpha  # общий коэффициент
        self.l1_ratio = l1_ratio  # пропорция L1 / L2
        self.learning_rate = learning_rate
        self.tolerance = tolerance
        self.max_iter = max_iter

    def fit(self, X, y):
        X_poly = polynomial_features(X, self.degree)
        n_samples, n_features = X_poly.shape
        self.weights = np.zeros(n_features)

        l1 = self.alpha * self.l1_ratio
        l2 = self.alpha * (1 - self.l1_ratio)

        for _ in range(self.max_iter):
            y_pred = X_poly @ self.weights
            error = y_pred - y

            l1_grad = np.sign(self.weights)
            l1_grad[0] = 0
            l2_grad = np.r_[0, self.weights[1:]]

            gradient = (1/n_samples) * X_poly.T @ error + l1 * l1_grad + l2 * l2_grad
            new_weights = self.weights - self.learning_rate * gradient

            if np.all(np.abs(new_weights - self.weights) < self.tolerance):
                break

            self.weights = new_weights

    def predict(self, X):
        X_poly = polynomial_features(X, self.degree)
        return X_poly @ self.weights
```

```python
degree = 2
alpha = 10  # сила регуляризации (настраивается)
l1_ratio = 0.5  # ElasticNet: 0 = Ridge, 1 = Lasso

# --- Обучение моделей ---
ridge = PolynomialRidgeGDRegression(degree=degree, alpha=alpha)
ridge.fit(X, y)
ridge_preds = ridge.predict(X)

lasso = PolynomialLassoGDRegression(degree=degree, alpha=alpha)
lasso.fit(X, y)
lasso_preds = lasso.predict(X)

enet = PolynomialElasticNetGDRegression(degree=degree, alpha=alpha, l1_ratio=l1_ratio)
enet.fit(X, y)
enet_preds = enet.predict(X)

# --- Визуализация ---
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Actual Data')
plt.plot(X, ridge_preds, color='blue', linestyle='--', label='Ridge')
plt.plot(X, lasso_preds, color='green', linestyle='-.', label='Lasso')
plt.plot(X, enet_preds, color='red', linestyle='-', label='ElasticNet')
plt.title(f'Ridge vs Lasso vs ElasticNet (degree={degree}, alpha={alpha})')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# --- Метрики ---
def print_metrics(name, y_true, y_pred):
    print(f'{name}:\n  MSE = {mean_squared_error(y_true, y_pred):.3f}\n  R²  = {r2_score(y_true, y_pred):.3f}\n')

print_metrics("Ridge", y, ridge_preds)
print_metrics("Lasso", y, lasso_preds)
print_metrics("ElasticNet", y, enet_preds)
```

Если alpha маленький (0.01, 0.1) — все модели будут похожи на обычную линейную регрессию.

Если alpha большой (10, 100) — Lasso может обнулить какие-то веса, Ridge сгладит кривую, ElasticNet — нечто среднее.

Если добавить шум — регуляризаторы должны уменьшить переобучение (более сглаженные кривые).

## Преимущества и недостатки линейной регрессии

#### Преимущества:

простота в реализации и интерпретации;

высокая скорость работы;

относительно хорошая точность в случае с линейной зависимостью в данных.

#### Недостатки:

низкая гибкость и адаптивность из-за предположения о линейности данных;

низкая точность в случае с данными сложной формы, что следует из предыдущего пункта;

чувствительность к шуму и выбросам.
