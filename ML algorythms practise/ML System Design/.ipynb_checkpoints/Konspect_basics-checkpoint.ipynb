# Общая картина

ML-система — это не просто модель, а цепочка процессов от данных до готового сервиса, которая включает:

1) Training pipeline — как мы готовим модель.
2) Serving pipeline — как модель обслуживает запросы в реальном мире.
3) Мониторинг — как мы следим, чтобы модель не деградировала.

# Training Pipeline

Цель — превратить «сырые» данные в обученную модель.

### Этапы:

1) Сбор данных — базы, API, логи, сенсоры.
2) Предобработка — очистка, кодирование категорий, нормализация.
3) Фичеинжиниринг — создание новых признаков.
4) Разделение — train / validation / test.
5) Обучение и тюнинг — Grid Search, Optuna, AutoML.
6) Валидация — оценка по метрикам (accuracy, RMSE, ROC AUC).
7) Версионирование — сохранение данных и модели (DVC, MLflow).

```python
# training_pipeline.py
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib

# 1. Сбор данных
X, y = load_iris(return_X_y=True)

# 2. Разделение
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 3. Обучение
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# 4. Сохранение модели
joblib.dump(clf, "model.pkl")
```

# Serving Pipeline

Цель — принимать входные данные и выдавать предсказания.

### Варианты:

* Batch inference — периодическая обработка больших пакетов данных.
* Online inference — предсказание для каждого запроса в реальном времени.

Пример (FastAPI + модель):

```python
# serving_pipeline.py
from fastapi import FastAPI
import joblib
import numpy as np

app = FastAPI()
model = joblib.load("model.pkl")

@app.post("/predict")
def predict(features: list):
    prediction = model.predict(np.array([features]))
    return {"prediction": prediction.tolist()}
```

```python
Запуск:

uvicorn serving_pipeline:app --reload
```

```python
Тест:

curl -X POST "http://127.0.0.1:8000/predict" -H "Content-Type: application/json" -d "[5.1, 3.5, 1.4, 0.2]"
```

### API протоколы

* REST — простой, JSON, HTTP.
* gRPC — бинарный, быстрее, для high-load.
* WebSocket — для потоковых данных (чат, стриминг).

# Мониторинг

### Что мониторить:

1) Data Drift — входные данные отличаются от тех, на которых обучали модель.
2) Model Drift — модель стала хуже работать (метрики падают).
3) Инфраструктура — задержка, ошибки, нагрузка.

Пример мониторинга дрейфа (Evidently AI):

```python
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset
import pandas as pd

# Исторические данные (train)
reference = pd.read_csv("train_data.csv")
# Новые данные (production batch)
current = pd.read_csv("new_data.csv")

report = Report(metrics=[DataDriftPreset()])
report.run(reference_data=reference, current_data=current)
report.save_html("drift_report.html")
```

# CI/CD для ML

* CI (Continuous Integration) — автоматические тесты и валидация модели.
* CD (Continuous Delivery/Deployment) — автоматическая доставка обновлений в прод.
* Инструменты: GitHub Actions, GitLab CI, Jenkins.

### Мини-пайплайн:

1) Новый датасет → запуск тренировки → тестирование метрик.
2) Если метрика лучше предыдущей → пуш в репозиторий моделей.
3) Автодеплой в API-сервис.
